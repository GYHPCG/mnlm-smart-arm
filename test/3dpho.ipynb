```python
import matplotlib.pyplot as plt
import numpy as np

# Data
instruction_types = ['抓取小方块', '移动至指定位置', '装配组件', '旋转调整', '精确定位']
success_rates = [0.75, 0.65, 0.60, 0.80, 0.70]
partial_success = [0.15, 0.20, 0.25, 0.10, 0.20]
failure_rates = [0.10, 0.15, 0.15, 0.10, 0.10]

# Plot setup
x_indexes = np.arange(len(instruction_types))
bar_width = 0.6

# Create stacked bars
plt.figure(figsize=(10, 6))
plt.bar(x_indexes, success_rates, width=bar_width, label='成功', color='green')
plt.bar(x_indexes, partial_success, width=bar_width, bottom=success_rates, label='部分成功', color='gold')
plt.bar(x_indexes, failure_rates, width=bar_width, bottom=np.array(success_rates) + np.array(partial_success), label='失败', color='red')

# Axis labels and title
plt.xlabel('指令类型', fontsize=12)
plt.ylabel('成功率 (%)', fontsize=12)
plt.title('机械臂指令执行效果分析', fontsize=14)
plt.xticks(x_indexes, instruction_types)
plt.yticks(np.arange(0, 1.1, 0.1), ['{:,.0%}'.format(y) for y in np.arange(0, 1.1, 0.1)])
plt.ylim(0, 1)

# Legend and grid
plt.legend(loc='upper right')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Save as high-resolution image
plt.tight_layout()
plt.savefig('mechanical_arm_performance.png', dpi=300, bbox_inches='tight')
plt.close()
```
